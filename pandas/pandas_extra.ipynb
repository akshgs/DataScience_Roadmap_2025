# loc and ilco
### loc (Label-based indexing)
Uses row labels (index names) and column names.
### iloc (Integer-location based indexing)
Uses integer positions (like Python lists)

```python
df
```

```python
df.loc[0]
```

```python
df.loc[0:5,['Day','Units Sold']]
```
