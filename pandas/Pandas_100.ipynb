## 82. Write a custom aggregation function inside .agg().

```python
df=pd.DataFrame({
    'team':['A','A','B','B','B'],
    'Score':[10,15,7,9,12]
})
def data_range(s: pd.Series) -> float:
    """Return max - min of a Series (works as an agg)."""
    return s.max() - s.min()

result = df.groupby("team").agg(score_range=("Score", data_range))
print(result)
    
```
