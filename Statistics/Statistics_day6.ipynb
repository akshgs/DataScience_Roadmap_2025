## ðŸ”¹ 1ï¸âƒ£ What is Hypothesis Testing?

**Definition:**  
Hypothesis testing is a **statistical method** used to make **decisions or inferences about a population** based on sample data.

In simple terms â€” it helps us test whether the result we see in the sample is **due to chance** or represents a **real effect** in the population.

---

### **Example Scenario**
A data scientist wants to test if a new marketing strategy increases average sales.  
- **Old average sales (Âµâ‚€)** = $5000  
- **New sample average** = $5300  

Now, the question is:  
> â€œIs the increase from $5000 to $5300 significant or just due to random variation?â€

This is where **hypothesis testing** is used.

## ðŸ”¹ 2ï¸âƒ£ Components of Hypothesis Testing

| **Component** | **Meaning** | **Example** |
|----------------|-------------|--------------|
| **Population Parameter (Âµ, p)** | The actual mean or proportion in the population | True mean weight of products |
| **Sample Statistic (xÌ„, pÌ‚)** | The calculated mean or proportion from the sample | Sample mean from 100 items |
| **Test Statistic** | A standardized value (like z, t) used to compare sample and population | z-test, t-test |
| **Significance Level (Î±)** | The probability of rejecting a true null hypothesis (Type I error) | Common values: 0.05 or 0.01 |
| **p-value** | Probability of obtaining the observed result if the null hypothesis is true | p = 0.02 indicates strong evidence against Hâ‚€ |

## ðŸ”¹ 3ï¸âƒ£ Null and Alternative Hypotheses

| **Type** | **Symbol** | **Meaning** | **Example** |
|-----------|-------------|--------------|--------------|
| **Null Hypothesis** | \( H_0 \) | Assumes **no effect or no difference**; the default assumption | \( H_0: \mu = 5000 \) (mean sales = 5000) |
| **Alternative Hypothesis** | \( H_1 \) or \( H_a \) | Claims there **is an effect or difference** | \( H_1: \mu > 5000 \) (mean sales increased) |

---

### **Types of Alternative Hypotheses**
| **Type** | **Condition** | **Example** |
|-----------|----------------|--------------|
| **Right-tailed test** | \( H_1: \mu > \mu_0 \) | Testing if new mean > old mean |
| **Left-tailed test** | \( H_1: \mu < \mu_0 \) | Testing if new mean < old mean |
| **Two-tailed test** | \( H_1: \mu \neq \mu_0 \) | Testing if new mean â‰  old mean |

---

## ðŸ”¹ 4ï¸âƒ£ Understanding p-value and Significance Level (Î±)

### **p-value**
- The **probability of observing** your sample results (or more extreme) if **Hâ‚€ is true**.  
- **Small p-value â†’ strong evidence against Hâ‚€.**

| **p-value** | **Interpretation** |
|--------------|------------------|
| p > 0.05 | Weak evidence against Hâ‚€ â†’ Fail to reject Hâ‚€ |
| p â‰¤ 0.05 | Strong evidence against Hâ‚€ â†’ Reject Hâ‚€ |
| p â‰¤ 0.01 | Very strong evidence against Hâ‚€ |

---

### **Significance Level (Î±)**
- The **threshold** we set to decide whether to reject Hâ‚€.  
- Common choices: **0.05 (5%)** or **0.01 (1%)**

**Decision Rule:**  
\[
\text{If } p \leq \alpha \Rightarrow \text{Reject } H_0  
\]  
\[
\text{If } p > \alpha \Rightarrow \text{Fail to Reject } H_0
\]

## ðŸ”¹ 5ï¸âƒ£ Steps in Hypothesis Testing

| **Step** | **Description** | **Example** |
|-----------|----------------|--------------|
| **1. State Hypotheses** | Define Hâ‚€ and Hâ‚ | \( H_0: \mu = 5000 \), \( H_1: \mu > 5000 \) |
| **2. Set Significance Level (Î±)** | Choose 0.05 or 0.01 | Î± = 0.05 |
| **3. Select Test Statistic** | Choose z-test, t-test, etc. | z-test for large sample |
| **4. Compute Test Statistic & p-value** | From sample data | p = 0.03 |
| **5. Make Decision** | Compare p with Î± | Since 0.03 < 0.05 â†’ Reject Hâ‚€ |
| **6. Conclusion** | Interpret in context | Sales increased significantly |

## ðŸ”¹ 6ï¸âƒ£ Errors in Hypothesis Testing

|--------------------|-------------|--------------|
| **Type I Error (Î±)** | Rejecting a true Hâ‚€ | Concluding sales increased when they didnâ€™t |
| **Type II Error (Î²)** | Failing to reject a false Hâ‚€ | Concluding no change when sales actually increased |
