---

### **Types of Variables**

| **Type** | **Definition** | **Example** | **Use in Data Science** |
|-----------|----------------|--------------|--------------------------|
| **Independent Variable** | The variable that is changed or controlled in an experiment. It influences the dependent variable. | Age, Study Hours, Ad Spend | Used as **input features (X)** in predictive models |
| **Dependent Variable** | The variable being tested or measured. It depends on the independent variable. | Exam Score, Sales, Weight | Used as **target/output (Y)** in models |
| **Continuous Variable** | Can take **any numeric value** within a range (including fractions/decimals). | Height = 5.8 ft, Temperature = 36.7¬∞C | Useful in **regression models** and numerical analysis |
| **Discrete Variable** | Can take only **specific or countable values** (no fractions). | Number of students = 50, Cars = 12 | Used in **classification and counting problems** |


### **Quick Summary**
- **Independent variable ‚Üí Input (X)**  
- **Dependent variable ‚Üí Output (Y)**  
- **Continuous variable ‚Üí Infinite possible values**  
- **Discrete variable ‚Üí Countable whole number

# Basic of Probability
**Definition:**
probbability measure the likelyhood of an event occuring
- 0 ‚Üí Impossible event  
- 1 ‚Üí Certain event  

\[
P(Event) = \frac{\text{Number of favorable outcomes}}{\text{Total number of outcomes}}
\]

**Example:**  
Rolling a dice ‚Üí Probability of getting even number = 3/6 = 0.5

---

### **Types of Probability**

| **Type** | **Definition** | **Example** |
|-----------|----------------|--------------|
| **Theoretical Probability** | Based on mathematical reasoning or formulas. | Probability of getting a head = 1/2 |
| **Experimental Probability** | Based on actual experiments or collected data. | Toss coin 100 times, get 52 heads ‚Üí 52/100 = 0.52 |
| **Subjective Probability** | Based on personal judgment or experience. | Chance of rain tomorrow = 80% (by meteorologist) |

---

### **Basic Probability Rules**

| **Rule** | **Formula** | **Meaning / Example** |
|-----------|-------------|------------------------|
| **Complement Rule** | \( P(A') = 1 - P(A) \) | If P(rain) = 0.3 ‚Üí P(no rain) = 0.7 |
| **Addition Rule** | \( P(A \cup B) = P(A) + P(B) - P(A \cap B) \) | Probability of A or B |
| **Multiplication Rule** | \( P(A \cap B) = P(A) \times P(B) \) (if independent) | Probability of A and B both occur |
| **Conditional Probability** | \( P(A|B) = \frac{P(A \cap B)}{P(B)} \) | Probability of A given B occurred |

---

### **Events in Probability**

| **Type of Event** | **Description** | **Example** |
|--------------------|----------------|--------------|
| **Independent Events** | One event does not affect the other. | Rolling two dice |
| **Dependent Events** | One event influences the other. | Drawing cards without replacement |
| **Mutually Exclusive Events** | Both cannot happen together. | Getting head **or** tail in one toss |
| **Exhaustive Events** | All possible outcomes considered. | Outcomes on a dice: 1‚Äì6 |
