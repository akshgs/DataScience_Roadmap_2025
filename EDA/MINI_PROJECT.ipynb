### Outlier Detection

```python
sns.boxplot(df)
plt.xticks(rotation=90)
plt.show()
```

- some outliers we can handle it drop or capping
  

```python
def outlier_detection(df, columns):
    outliers = {}
    for column in columns:
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr
        
        outlier = df[(df[column] < lower) | (df[column] > upper)][column]

        print(f'\nColumn: {column}')
        print(f'Lower Limit: {lower:.2f}')
        print(f'Upper Limit: {upper:.2f}')
        print(f'Outlier Count: {len(outlier)}')
        print(f'Outlier Percentage: {len(outlier) / len(df) * 100:.2f}%')
        print(f'Outlier Values: {list(outlier[:10])} ...')

        outliers[column] = outlier

    return outliers
```

```python
outlier_detection(df,['Age'])
```

- This is student addiction so lower limit=5 percentile,upper limit=95 percentile it is the basic range of student age
- The best way outlier handling is capping

```python
low = np.percentile(df['Age'],5,method='midpoint')
upp = np.percentile(df['Age'],95,method='midpoint')
df['Age'] = df['Age'].clip(lower=low,upper=upp)
```

 ### Addiction_Score into a categorical column Addiction_Level based on score ranges
**Purpose:**
To classify students by addiction intensity for easier analysis and visualization.

```python
def map_addiction_level(score):
    if score <= 2.5:
        return 'Low'
    elif 2.5 < score <= 4.0:
        return 'Moderate'
    else:
        return 'High'

df['Addiction_Level'] = df['Addiction_Score'].apply(map_addiction_level)
```

###  **Key Findings**

1. **High Screen Time → Higher Addiction:**

3. **Stress and Mental Health:**

4. **Academic Impact:**  
   Students with high addiction and stress levels perform worse academically, averaging 10–15 marks lower than low-addiction users.

5. **Platform Preference:**  
   Visual platforms like **TikTok** and **Instagram** dominate among high-addiction users, showing engagement patterns linked to shorter, immersive content.
