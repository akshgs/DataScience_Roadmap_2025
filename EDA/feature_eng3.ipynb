```python
def detect_outliers(df, columns):
    outlier_dict = {}

    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        outliers = df[(df[column] < lower) | (df[column] > upper)][column]

        print(f"\nColumn: {column}")
        print(f"Lower Limit: {lower}")
        print(f"Upper Limit: {upper}")
        print(f"Outlier Count: {len(outliers)}")
        print(f"Outlier Percentage: {round((len(outliers) / len(df)) * 100, 2)}%")
        print("Outlier Values:", list(outliers))

        outlier_dict[column] = outliers  

    return outlier_dict
```
