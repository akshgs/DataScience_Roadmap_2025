```python
num_data = new_df.select_dtypes(include=['int64', 'float64'])
num_data.head()


```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=num_data)
plt.title("Boxplot for Outlier Detection (All Numeric Columns)")
plt.xticks(rotation=45)
plt.show()
```

```python
Q1 = new_df['MonthlyCharges'].quantile(0.25)
Q3 = new_df['MonthlyCharges'].quantile(0.75)
IQR = Q3 - Q1
lowerlimit = Q1 - 1.5 * IQR
upperlimit = Q3 + 1.5 * IQR

outlier = new_df[(new_df['MonthlyCharges'] < lowerlimit) | (new_df['MonthlyCharges'] > upperlimit)]
print(outlier)
```
