```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.linspace(0,5,10)
y=x**2
plt.figure(figsize=(5,4))
plt.plot(x,y)
plt.title('Days Squred Chart')
plt.xlabel('Days')
plt.ylabel('Day Squre')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
y1 = [1, 3, 5, 7, 9]
plt.plot(x, y, label='Line 1')
plt.plot(x, y1, label='Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Example with Legend')
plt.legend()   
plt.show()
```
