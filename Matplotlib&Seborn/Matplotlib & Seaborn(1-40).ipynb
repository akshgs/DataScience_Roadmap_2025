## 15. Plot sine and cosine waves on the same axes

## 19. Show bar chart in one subplot and line plot in another.

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 9]
x = np.arange(1, 11)
y = x ** 2

plt.figure(figsize=(8,6))
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line plot')
plt.xlabel("Xaxis")
plt.ylabel("yaxis")
plt.subplot(1,2,2)
plt.bar(categories,values)
plt.title('Barplot')
plt.xlabel("Xaxis")
plt.ylabel("yaxis")
plt.show()
```

## 20. Display multiple lines with different styles and colors.

```python
x=np.linspace(1,10,100)
y=np.sin(x)
y1=np.cos(x)
y2=np.sin(x)*np.cos(x)

plt.plot(x,y,color='red',marker='o',label='sin x')
plt.plot(x,y1,color='blue',linestyle='--',marker='s',label='cos x')
plt.plot(x,y2,color='green',linestyle='dashdot',marker='^',label='sin x * cos x')
plt.legend()
plt.show()
```

## Use plt.style.use('ggplot') to apply a theme.

```python
x=np.linspace(1,10,100)
y=np.sin(x)
y1=np.cos(x)
y2=np.sin(x)*np.cos(x)

plt.style.use('ggplot')

plt.plot(x,y,color='red',marker='o',label='sin x')
plt.plot(x,y1,color='blue',linestyle='--',marker='s',label='cos x')
plt.plot(x,y2,color='green',linestyle='dashdot',marker='^',label='sin x * cos x')
plt.legend()
plt.title('Multiple Lines with ggplot Style')
plt.show()
```

## 24. Add a vertical line (axvline) to indicate a threshold.

```python
x=np.linspace(1,10,100)
y=np.sin(x)

plt.plot(x,y,color='red',label='X sine Value')
plt.axvline(x=5,color='pink',linestyle='--',linewidth='2',label='Thershold')
plt.legend()
plt.show()
```

## 25. Annotate the maximum point on a line chart.

```python
x=np.linspace(0,2*np.pi,100)
y=np.sin(x)

plt.plot(x,y,color='blue',label='Sin(x)')
max_index=np.argmax(x)
max_x=x[max_index]
max_y=y[max_index]

plt.scatter(max_x,max_y,color='red')

plt.annotate(f'Max: ({max_x:.2f}, {max_y:.2f})',
             xy=(max_x, max_y), 
             xytext=(max_x + 0.5, max_y + 0.1),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Chart with Maximum Point Annotated')
plt.legend()
plt.grid(True)
```

## 39. Create stacked bar charts

```python
subjects = ['Math', 'Physics', 'Chemistry', 'Biology', 'English']
marks_A = [50, 40, 60, 70, 45]   
marks_B = [30, 35, 25, 20, 40]
marks_C = [20, 25, 15, 10, 15]
x=np.arange(len(subjects))
plt.bar(x,marks_A,color='skyblue',label='Class A')
plt.bar(x,marks_B,color='lightgreen',label='Class B')
plt.bar(x,marks_C,color='orange',bottom=np.array(marks_A)+np.array(marks_B),label='Class C')
plt.xticks(x,subjects)
plt.xlabel('Subjects')
plt.ylabel('Mark')
plt.legend()
plt.show()
```

## 40. Plot cumulative line using cumsum().
