```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 56. Load the tips dataset and display the first few rows.

```python
df=sns.load_dataset('tips')
df.head(5)
```

## 57. Plot total bill vs tip using sns.scatterplot.

```python
sns.scatterplot(df,x='total_bill',y='tip')
plt.show()
```

## 58. Color points by sex column.

```python
sns.scatterplot(df,x='total_bill',y='tip',hue='sex',palette='coolwarm',s=100)
plt.title('Scetter plot')
plt.show()
```

## 59. Add regression line using sns.lmplot.

```python
sns.lmplot(df,x='total_bill',y='tip',hue='sex',markers=['o','s'])
plt.title('Regression line by sex')
plt.show()
```

## 60. Plot a histogram using sns.histplot.

```python
sns.histplot(df,x='total_bill',bins=2)
plt.title('hist plot sex')
plt.show()
```

## 61. Plot a KDE (density) plot using sns.kdeplot

```python
sns.kdeplot(df,x='total_bill')
plt.show()
```

## 62. Create a barplot showing average tip by day.

```python
avg=df.groupby('day',observed=True)['tip'].mean().reset_index()
sns.barplot(avg,x='day',y='tip')
plt.show()
```

## 63. Use estimator='sum' in sns.barplot.

```python
sns.barplot(df,x='day',y='tip',estimator='sum')
plt.show()
```

## 64. Create a countplot showing number of observations per day.

```python
sns.countplot(df,x='day',hue='day',palette='viridis')
plt.title('count plot of each day')
plt.show()
```

## 65. Plot a boxplot for total_bill grouped by day.

```python
sns.boxplot(df,x='day',y='total_bill')
plt.title('total_bill grouped by day')
plt.xlabel('Day')
plt.show()
```

## 66.Plot a violin plot comparing tips by gender.

```python
sns.violinplot(df,x='sex',y='tip')
```

## 67. Add hue to your boxplot.

```python
sns.boxplot(df,y='tip',hue='sex')
plt.show()
```

## 68. Create a swarmplot for day vs tip.

```python
sns.swarmplot(df,x='day',y='tip')
plt.show()
```

## 69. Combine boxplot and swarmplot together.

```python
plt.figure(figsize=(8,6))
sns.boxplot(df,y='tip',hue='sex',palette='Set2')
sns.swarmplot(df,y='tip',hue='sex',dodge=True,color='0.25')
```

## 70. Draw a pairplot for all numeric columns in tips.

```python
sns.pairplot(df)
plt.show()
```

## 71. Create a heatmap for correlations of numeric columns.

```python
n=df.select_dtypes(include='number')
sns.heatmap(n,cmap='viridis')
plt.title('heatmap of all numeric value')
plt.show()
```

## 72. Use annot=True in heatmap.

```python
n=df.select_dtypes(include='number')
sns.heatmap(n,annot=True,cmap='viridis')
plt.title('heatmap of all numeric value')
plt.show()
```
