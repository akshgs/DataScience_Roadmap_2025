## 41.Create a 2Ã—2 grid of subplots

```python
x=np.linspace(1,10,100)
y=np.cos(x)
y1=np.sin(x)
plt.figure(figsize=(4,2))
plt.subplot(1,2,1)
plt.title('Cos wave')
plt.plot(x,y,color='black',linestyle='--')
plt.subplot(1,2,2)
plt.title('Sine Wave')
plt.plot(x,y1,color='lightgreen')

plt.show()
```

## 42. Share x-axis between subplots

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig,(ax1,ax2)=plt.subplots(2,1,sharex=True,figsize=(8,6))

ax1.plot(x, y1, color='blue')
ax1.set_ylabel('sin(x)')
ax1.set_title('Shared X-axis Example')

ax2.plot(x,y2,color='red')
ax2.set_xlabel('cos(x)')
ax2.set_ylabel('cos')

plt.tight_layout()
plt.show()
```

## 44. Adjust space between subplots with plt.subplots_adjust.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2, 1, figsize=(8,6))

axs[0].plot(x, y1)
axs[0].set_title("Sine Wave")

axs[1].plot(x, y2)
axs[1].set_title("Cosine Wave")

# Main title for all subplots
fig.suptitle("Main Title for All Subplots", fontsize=16)

# Adjust space between subplots
plt.subplots_adjust(
    top=0.88,   # space from top (for main title)
    bottom=0.1, # bottom margin
    left=0.1,   # left margin
    right=0.9,  # right margin
    hspace=0.4, # vertical space between subplots
    wspace=0.3  # horizontal space (if multiple columns)
)

plt.show()
```

## 48. Combine line and bar plots in one figure using two axes (twinx).

```python
data=['apple','orange','banana','mango','pappaya']
count=np.random.randint(6,22,5)
quli=np.random.randint(1,10,5)
fig,ax1=plt.subplots(figsize=(8,4))
bar = ax1.bar(data, count, color='skyblue', edgecolor='black', label='Count')
ax1.set_xlabel('Fruits')
ax1.set_ylabel('Count', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2=ax1.twinx()
line=ax2.plot(data,quli,color='red',marker='o',label='quality')
ax2.set_ylabel('Quality',color='red')
ax2.tick_params(axis='y',labelcolor='red')

handles = [bar, line[0]]
labels = ['Count', 'Quality']
ax1.legend(handles, labels, loc='upper left')

plt.title('Count vs Quality for Fruits (Bar + Line with twinx)')
plt.tight_layout()
plt.show()
```

## 52. Add an inset zoom plot inside a larger plot.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Sample data
x = np.linspace(0, 10, 500)
y = np.sin(x) + 0.2*np.random.randn(500)  # sine wave with noise

# Main plot
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, color='blue')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Main Plot with Zoom Inset')

# Add an inset axes (width=30%, height=30% of the main plot)
axins = inset_axes(ax, width="30%", height="30%", loc='upper right')

# Plot the same data in the inset
axins.plot(x, y, color='red')

# Set limits for zoom region
x1, x2 = 2, 4   # zoomed x-range
y1, y2 = -1, 1  # zoomed y-range
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)

# Optional: remove ticks for clarity
axins.set_xticks([])
axins.set_yticks([])

# Draw a rectangle on the main plot showing the zoom area
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="0.5")

plt.show()
```
