```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 76. Customize Seaborn theme using sns.set_style().

```python
df=sns.load_dataset('tips')
sns.set_style('whitegrid')
sns.lineplot(df,x='total_bill',y='tip')
plt.show()
```

## 77. Use sns.displot with multiple subplots (col and row).

```python
sns.displot(df,x='total_bill',col='time',row='sex',bins=15,kde=True,height=4,aspect=1.2)

plt.suptitle('Distribution of Total Bill by Sex and Time', y=1.03)
plt.show()
```

## 78. Add col_wrap to layout multiple plots.

```python
sns.displot(df,x='total_bill',col='time',bins=15,kde=True,height=4,aspect=1.2,col_wrap=2)

plt.suptitle('Distribution of Total Bill by Sex and Time', y=1.03)
plt.show()
```

## 79. Use FacetGrid for advanced grouping plots

```python
g=sns.FacetGrid(df,col='day')
g.map(sns.histplot,'total_bill',bins=15,color='skyblue')



plt.suptitle('Total Bill vs Tip by Sex and Time', y=1.05)
plt.show()
```

## 80. Plot multiple regression lines using sns.lmplot(hue='sex').

```python
sns.lmplot(data=df, x='total_bill', y='tip', hue='sex')
```

## 81. se sns.catplot(kind='bar').

```python
sns.catplot(df,x='sex',y='tip',kind='bar',legend=False,palette='pastel')
plt.legend(title='Gender',labels=['male','female'])
plt.show()
```

## 82. Create a jointplot for total_bill vs tip.

```python
sns.jointplot(df,x='total_bill',y='tip')
plt.show()
```

## 83. Display hex bins in sns.jointplot(kind='hex').

```python
sns.jointplot(
    data=df,
    x='total_bill',
    y='tip',
    kind='hex',          
    gridsize=20,         
    cmap='viridis'      
)
plt.show()
```

## 84. Use sns.stripplot for category scatter visualization.

```python
sns.stripplot(df,x='day',y='tip')
plt.show()
```

## 85. Create a boxen plot to visualize large data distribution.

```python
sns.boxenplot(df,x='tip',y='time')
plt.show()
```

## 86. Draw multiple heatmaps side by side for comparison

```python
corr=df.corr(numeric_only=True)
cormale=df[df['sex']=='Male'].corr(numeric_only=True)

fig, axes=plt.subplots(1,2,figsize=(12,5))
sns.heatmap(corr,annot=True,cmap='coolwarm',ax=axes[0])
axes[0].set_title('All data')

sns.heatmap(cormale,annot=True,ax=axes[1])
axes[1].set_title('Male Data')

plt.suptitle('Multiple heatmap')
plt.show()
```

## 87. Combine matplotlib titles with seaborn figures.

```python
tips = sns.load_dataset('tips')

# Create a seaborn plot
sns.scatterplot(data=tips, x='total_bill', y='tip')

# Add matplotlib titles and labels
plt.title('Relationship between Total Bill and Tip', fontsize=14)
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')

plt.show()
```

## 88. Change Seaborn context to 'talk' or 'poster'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

with sns.plotting_context("talk"):
    sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', s=100)
    plt.title("Tips vs Total Bill (Talk Context)")
    plt.show()
```

## 90. Customize figure size using height and aspect in Seaborn plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    height=6,     # subplot height in inches
    aspect=1.5   # width = 1.5 * height
)

plt.title('Total Bill vs Tip by Sex')
plt.show()
```
